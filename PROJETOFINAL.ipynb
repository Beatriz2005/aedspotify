# **Existem msicas repetidas?**




```python
#excluindo possveis dados duplicados, considerando msicas com mesmo nome, artista e popularidade, j que podem existir diferentes verses da mesma msica
dados_sem_repetir = dados.drop_duplicates(subset=['track_name', 'artists', 'popularity'], keep='first')
```

```python
dados_sem_repetir
```

# **Estudo da varivel *popularidade***

```python
#Verificando valores nulos
dados_sem_repetir["popularity"].isnull().sum()
```

Verificando se existem fora da faixa esperada [0, 100]

```python
valores_extremos = dados_sem_repetir[dados['popularity'] > 100 ]
print(valores_extremos)
```

```python
valores_abaixo = dados_sem_repetir[dados_sem_repetir['popularity'] < 0 ]
print(valores_abaixo)
```

```python
#Viso geral da varivel popularidade
dados_sem_repetir['popularity'].describe()
```

```python
dados_sem_repetir['popularity'].mean()
```

```python
dados_sem_repetir['popularity'].median()
```

Identificando possveis outliers:

```python
plt.figure(figsize=(10, 6))
sns.set(style="darkgrid")
sns.boxplot(x=dados_sem_repetir['popularity'])
plt.title('Popularidade')
plt.show()
```

```python
z_score = zscore(dados_sem_repetir['popularity'])
outliers_indices = (z_score > 3) | (z_score < -3)
outliers_musica = dados_sem_repetir[outliers_indices]
print("Msicas com popularidade considerada outlier:")
print(outliers_musica)
```

```python
outliers_nomes = dados_sem_repetir.loc[outliers_indices, ['track_name', 'popularity']]
print("Nomes das msicas com popularidade considerada outlier:")
print(outliers_nomes)
```

```python
outliers_nomes.shape
```

Apesar de no existirem erros nos dados, j que 100  um valor possvel para a varivel popularidade. Percebe-se que existem algumas msicas superpopulares (com popularidade acima ou igual a 95), porm isso  raro na grande variedade de popularidade existente do conjunto de dados, o que faz com que sejam outliers.

Visualizando a distribuio dos valores na varivel:

```python
fig, ax = plt.subplots(figsize = (10, 6))
sns.set_style("whitegrid")
ax.hist(dados_sem_repetir['popularity'], bins=20, edgecolor="black", color="#69b3a2", alpha= 1)
plt.xlabel('Popularidade')
plt.ylabel('Frequncia')
plt.title('Distribuio da Popularidade')
plt.show()
```

```python
top_1000_mais_populares = dados_sem_repetir.nlargest(1000, 'popularity')
# Verifica se existem valores nulos nas variveis de interesse
variaveis_interesse = ['danceability', 'valence', 'energy']
print(top_1000_mais_populares[variaveis_interesse].isnull().sum())

# Elimina linhas com valores nulos nas variveis de interesse
top1000 = top_1000_mais_populares.dropna(subset=variaveis_interesse)

# Cria os grficos de disperso
plt.figure(figsize=(12, 8))
sns.scatterplot(x='danceability', y='popularity', data=top_1000_mais_populares)
plt.title('Relao entre Danceabilidade e Popularidade nas Top 1000 Msicas')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(x='valence', y='popularity', data=top_1000_mais_populares)
plt.title('Relao entre Valncia e Popularidade nas Top 1000 Msicas')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(x='duration_ms', y='popularity', data=top_1000_mais_populares)
plt.title('Relao entre Durao e Popularidade nas Top 1000 Msicas')
plt.xlabel('Durao (ms)')
plt.show()

```

```python
correlacao_danceabilidade_popularidade = top_1000_mais_populares['danceability'].corr(top_1000_mais_populares['popularity'])
correlacao_valence_popularidade = top_1000_mais_populares['valence'].corr(top_1000_mais_populares['popularity'])

print(f'Correlao entre Danceabilidade e Popularidade: {correlacao_danceabilidade_popularidade}')
print(f'Correlao entre Valncia e Popularidade: {correlacao_valence_popularidade}')

correlacao_duracao_popularidade = top_1000_mais_populares['duration_ms'].corr(top_1000_mais_populares['popularity'])
print(f'Coeficiente de Correlao entre Durao e Popularidade: {correlacao_duracao_popularidade}')
```


## **Concluso**
Ao analisar as top 1000 msicas no Spotify, identificamos padres interessantes nas correlaes entre caractersticas musicais e sua popularidade. A danceabilidade mostra uma correlao positiva moderada, indicando que msicas mais propcias  dana tendem a ser mais populares. Por outro lado, a valncia, que expressa a positividade da msica, exibe uma correlao negativa leve, sugerindo que composies com atmosfera menos alegre podem ter uma tendncia a serem menos populares. Alm disso, a durao das msicas apresenta uma correlao negativa leve, indicando que faixas mais curtas tm uma propenso a alcanar maior popularidade. Essas observaes oferecem insights valiosos sobre as preferncias do pblico, ressaltando a complexidade das relaes entre caractersticas musicais especficas e atratividade para o ouvinte.
