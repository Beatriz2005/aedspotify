O conjunto de dados possui 19 atributos que representam características das músicas e 114000 tuplas. Os atributos são:

**artists:** nome do artista (string)

**album_name:** nome do álbum (string)

**track_name:** nome da música (String)

**popularity:** pontuação de popularidade, variando de 0 a 100 (inteiro)

**duration_ms:** a duração de música em milissegundos (inteiro)

**explicit:** se a músita em conteúdo explícito (boolean)

**danceability:**pontuação que varia de 0 a 1 que representa o quão adequada é uma faixa para dançar com base em vários elementos musicais. (float)

**energy:** medida da intensidade e da atividade de uma faixa, que varia de 0 a 1. (floar)

**key:** o tom da faixa representado por um valor inteiro. (Inteiro)

**loudness:** volume da faixa em decibéis (dB). (Float)

**mode:** modo tonal da faixa, representado por um valor inteiro (0 para menor, 1 para maior). (Inteiro)

**speechiness:** pontuação de 0 a 1 que representa a presença de palavras faladas numa faixa. (Float)

**acousticness:** pontuação de 0 a 1 que representa o grau de qualidade acústica de uma faixa. (Float)

**instrumentalness:** pontuação de 0 a 1 que representa a probabilidade de uma faixa ser instrumental. (float)

**liveness:** pontuação de 0 a 1 que representa a presença de um público durante a gravação ou a execução de uma faixa. (float)

**valence:** pontuação de 0 a 1 que representa a positividade musical transmitida por uma faixa. (float)

**tempo:** tempo da faixa em batidas por minuto (BPM). (float)

**time_signature:** número de batidas dentro de cada compasso da faixa. (Inteiro)

**track_genre:** o gênero da música

As outras 2 colunas:

**Unnamed**: índice das músicas

**track_id**: trata-se do id das músicas no Spotify

# **Existem músicas repetidas?**




```python
#excluindo possíveis dados duplicados, considerando músicas com mesmo nome, artista e popularidade, já que podem existir diferentes versões da mesma música
dados_sem_repetir = dados.drop_duplicates(subset=['track_name', 'artists', 'popularity'], keep='first')
```

```python
dados_sem_repetir
```

# **Estudo da variável *popularidade***

```python
#Verificando valores nulos
dados_sem_repetir["popularity"].isnull().sum()
```

Verificando se existem fora da faixa esperada [0, 100]

```python
valores_extremos = dados_sem_repetir[dados['popularity'] > 100 ]
print(valores_extremos)
```

```python
valores_abaixo = dados_sem_repetir[dados_sem_repetir['popularity'] < 0 ]
print(valores_abaixo)
```

```python
#Visão geral da variável popularidade
dados_sem_repetir['popularity'].describe()
```

```python
dados_sem_repetir['popularity'].mean()
```

```python
dados_sem_repetir['popularity'].median()
```

Identificando possíveis outliers:

```python
plt.figure(figsize=(10, 6))
sns.set(style="darkgrid")
sns.boxplot(x=dados_sem_repetir['popularity'])
plt.title('Popularidade')
plt.show()
```

```python
z_score = zscore(dados_sem_repetir['popularity'])
outliers_indices = (z_score > 3) | (z_score < -3)
outliers_musica = dados_sem_repetir[outliers_indices]
print("Músicas com popularidade considerada outlier:")
print(outliers_musica)
```

```python
outliers_nomes = dados_sem_repetir.loc[outliers_indices, ['track_name', 'popularity']]
print("Nomes das músicas com popularidade considerada outlier:")
print(outliers_nomes)
```

```python
outliers_nomes.shape
```

Apesar de não existirem erros nos dados, já que 100 é um valor possível para a variável popularidade. Percebe-se que existem algumas músicas superpopulares (com popularidade acima ou igual a 95), porém isso é raro na grande variedade de popularidade existente do conjunto de dados, o que faz com que sejam outliers.

Visualizando a distribuição dos valores na variável:

```python
fig, ax = plt.subplots(figsize = (10, 6))
sns.set_style("whitegrid")
ax.hist(dados_sem_repetir['popularity'], bins=20, edgecolor="black", color="#69b3a2", alpha= 1)
plt.xlabel('Popularidade')
plt.ylabel('Frequência')
plt.title('Distribuição da Popularidade')
plt.show()
```

## **Conclusão**

  A variável popularidade mostrou estar concentrada abaixo de 50, com média igual a 34.66 e mediana igual a 35.0.

# **Quais são os gêneros mais populares no conjunto de dados?**

```python
dados_sem_repetir["track_genre"].isnull().sum()
```

```python
dados_sem_repetir["track_genre"].nunique()
```

```python
genero_popularidade = dados_sem_repetir.groupby('track_genre')['popularity'].mean()
genero_popularidade = genero_popularidade.sort_values(ascending=False)
top_20_generos = genero_popularidade.head(20)
```

```python
top_20_generos
```

```python
plt.figure(figsize=(12, 8))
top_20_generos.plot(kind='barh', edgecolor="black", color='#69b3a2', width=0.8)
plt.xlabel('Popularidade Média')
plt.ylabel('Gênero')
plt.title('Top 20 Gêneros Mais Populares', fontsize=16)
plt.show()
```

```python
dados_sem_repetir['track_genre'] = pd.Categorical(dados_sem_repetir['track_genre']).codes
correlation = dados_sem_repetir['popularity'].corr(dados_sem_repetir['track_genre'])
print(f"Correlação entre popularidade e gênero: {correlation}")
```

## **Conclusão**
A correlação entre o gênero e a popularidade das músicas é de apenas 0.03436. Isso significa que há uma relação muito fraca ou quase nula entre essas duas coisas no conjunto de dados.

```python
top_1000_mais_populares = dados_sem_repetir.nlargest(1000, 'popularity')
# Verifica se existem valores nulos nas variáveis de interesse
variaveis_interesse = ['danceability', 'valence', 'energy']
print(top_1000_mais_populares[variaveis_interesse].isnull().sum())

# Elimina linhas com valores nulos nas variáveis de interesse
top1000 = top_1000_mais_populares.dropna(subset=variaveis_interesse)

# Cria os gráficos de dispersão
plt.figure(figsize=(12, 8))
sns.scatterplot(x='danceability', y='popularity', data=top_1000_mais_populares)
plt.title('Relação entre Danceabilidade e Popularidade nas Top 1000 Músicas')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(x='valence', y='popularity', data=top_1000_mais_populares)
plt.title('Relação entre Valência e Popularidade nas Top 1000 Músicas')
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(x='duration_ms', y='popularity', data=top_1000_mais_populares)
plt.title('Relação entre Duração e Popularidade nas Top 1000 Músicas')
plt.xlabel('Duração (ms)')
plt.show()

```

```python
correlacao_danceabilidade_popularidade = top_1000_mais_populares['danceability'].corr(top_1000_mais_populares['popularity'])
correlacao_valence_popularidade = top_1000_mais_populares['valence'].corr(top_1000_mais_populares['popularity'])

print(f'Correlação entre Danceabilidade e Popularidade: {correlacao_danceabilidade_popularidade}')
print(f'Correlação entre Valência e Popularidade: {correlacao_valence_popularidade}')

correlacao_duracao_popularidade = top_1000_mais_populares['duration_ms'].corr(top_1000_mais_populares['popularity'])
print(f'Coeficiente de Correlação entre Duração e Popularidade: {correlacao_duracao_popularidade}')
```


## **Conclusão**
Ao analisar as top 1000 músicas no Spotify, identificamos padrões interessantes nas correlações entre características musicais e sua popularidade. A danceabilidade mostra uma correlação positiva moderada, indicando que músicas mais propícias à dança tendem a ser mais populares. Por outro lado, a valência, que expressa a positividade da música, exibe uma correlação negativa leve, sugerindo que composições com atmosfera menos alegre podem ter uma tendência a serem menos populares. Além disso, a duração das músicas apresenta uma correlação negativa leve, indicando que faixas mais curtas têm uma propensão a alcançar maior popularidade. Essas observações oferecem insights valiosos sobre as preferências do público, ressaltando a complexidade das relações entre características musicais específicas e atratividade para o ouvinte.
